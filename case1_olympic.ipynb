```
data['Medal'] = data['Medal'].fillna('none')
data.groupby('Medal')['Age'].mean()
```

```
women = data[data['Sex']=='F']
sns.set(style="darkgrid")
plt.figure(figsize=(10, 5))
plt.tight_layout()
count = sns.countplot(x='Year',data = women, hue='Season')
count.set_xticklabels(count.get_xticklabels(), rotation=-90)
plt.savefig('women_count')
```

```
Sexratio = data.groupby(['Games','Sex'],as_index=False)['ID'].count()
Sexratio.rename(columns={'ID':'count'},inplace=True)
Sexratio.sort_values(['Games','Sex'])
Sexverse = Sexratio.copy()
Sexverse['Sex'] = Sexverse['Sex'].apply(lambda x:'M' if x=='F' else 'F')
ratio = Sexratio.merge(Sexverse,how='outer',on=['Games','Sex'])
ratio['count_y']=ratio['count_y'].fillna(0)
ratio['ratio']=ratio['count_y']/ratio['count_x']
ratio
```

```
rate = ratio[ratio['Sex']=='M']
rate = rate[['Games','ratio']]
rate['Year'] = rate['Games'].apply(lambda x:int(x[:4]))
rate['Season'] = rate['Games'].apply(lambda x:x[5:])
rate.drop('Games',inplace=True,axis = 1)
rate.reset_index(drop=True,inplace=True)
rate
```

```
plt.figure(figsize=(10, 5))
pic = sns.lineplot(x='Year',y='ratio',hue = 'Season',data=rate)
pic.set_title('Sex ratio')
plt.savefig('ratio')
```
